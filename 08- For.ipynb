# **While** e **For** Loops

Instru√ß√µes `while` e `for` permitem implementar *la√ßos* de execu√ß√£o nos programas, os chamados *loops* \*. S√£o blocos de instru√ß√µes que se repetem at√© satisfazer alguma condi√ß√£o com chegar a um determinado valor ou √≠ndice.

Os *loops* s√£o muito √∫teis para codificar situa√ß√µes repetitivas sem a necessidade de extensas linhas de c√≥digo que se repetem.

\* *n√£o confunda, √†s vezes tamb√©m empregamos o t√™rmo 'loop' para quando um programa, geralmente por erro, 'n√£o para'.*





# Exemplo
Elabore um programa que calcule e apresente o valor de S, dado por:

$$ùëÜ = \frac{1}{N} + \frac{2}{N-1} + \frac{3}{N-2} + ... + \frac{N-1}{2} + N$$

Sendo N fornecido pelo usu√°rio.


```python
# Seu c√≥digo
n = int(input("Digite o valor de n: "))
S = 0
denominador = n

for i in range(1,n+1,1):
    print("termo",i ,"=",i,"/",denominador)

    S = S + i/denominador
    denominador = denominador - 1

print("\nA soma √©", "%.2f" %S)
```

# Exerc√≠cio 1

Escreva um programa que apresente todos os √≠mpares de 1 at√© 99.

```python
#uma estrutura de repeti√ß√£o que faz basicamente o seguinte: o for vai criar a
#vari√°vel {i} que vai repetir o "print i" do 1 at√© 100 (ou seja, 99 vezes).
#Nisso, vai ser somado +2 na vari√°vel {i}, ent√£o ele vai do 1, 3, 5... pois
# toda vez q a estrutura se repete ele vai ser somado a dois
for i in range(1, 100, 2):
      print(i)
```

# Exerc√≠cio 2

Escreva um programa que apresente os n (1 <= n <= 20) primeiros termos da seguinte sequ√™ncia: 1, 4, 9, 16, 25, ...

```python
#a sequ√™ncia basicamente √© uma tabuada quadr√°tica e √© isso que a estrutura de
#repeti√ß√£o est√° reproduzindo: ele vai multiplicar os n√∫meros de 1 a 21 por
#eles mesmos e depois vai imprimir essa multiplica√ß√£o ao usu√°rio.
for i in range(1, 21):
  print(i * i)
```

# Exerc√≠cio 3

Fa√ßa um programa que calcule e apresente o fatorial de um n√∫mero inteiro e natural fornecido
pelo usu√°rio.

Exemplo: 5! = 5 x 4 x 3 x 2 x 1 = 120.
Por defini√ß√£o 0! = 1.

```python
#vari√°vel que vai armazenar o resultado dao fatorial
fatorial = 1

#n√∫mero inteiro do usu√°rio que vai ser o n√∫mero fatorado
num = int(input('Digite um n√∫mero inteiro: '))

#for que cria a vari√°vel {i} e que vai come√ßar pelo {num} e vai ir at√© 1,
#ou seja, vai
for i in range (num, 1, -1):
   fatorial *= i

#impress√£o do resultado do fatorial
print(fatorial)
```

# Exerc√≠cio 4

A convers√£o de graus Fahrenheit para Celsius √© obtida pela f√≥rmula:

$$ùê∂ = \frac{5}{9}(ùêπ ‚àí 32)$$

Escreva um programa que calcule e apresente uma tabela de graus Celsius em fun√ß√£o de graus Fahrenheit que variem de 50 a 150, de 1 em 1.


```python
#estrutura de repeti√ß√£o que vai repetir dos 10¬∞C at√© os 66¬∞C (66¬∞C pois √© o a
#convers√£o de 150 Fahrenheit)
for i in range(10, 67):

  #equa√ß√£o que faz a convers√£o dos celcius (vari√°vel i) para Fahrenheit
  f = (i * 9/5) + 32

  #impress√£o do valor dos celseius e o valor de fahrenheit
  print(f'Celsius: {i}¬∞  Fahrenheit: {f}')
```
