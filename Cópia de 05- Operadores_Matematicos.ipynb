# EXEMPLO

```python
import math

e = math.exp(1)
pi = math.pi

print('Este é o número de Euler: ', e)
print('Este é o número de Euler:  {:0.2f}'.format(e)) # Prefira este, Python 3
print('Este é o número de Euler:  {:0.4f}'.format(e))

print ('Este é o número de Euler: ', format(e, '.4f'))
print ('Este é o número de Euler: ', '%.4f' %e)

print('Este é o número de Euler:  {:0.4f} , e este é o Pi: {:0.4f}'.format(e,pi))
```

# Exercício 1.

Escreva um programa que lê dois valores **inteiros** e exibe os seguintes resultados:

* o resultado do primeiro número dividido pelo segundo número
* o resultado da divisão truncada do primeiro número pelo segundo número


```python
#input do primeiro número do usuário
num1 = int(input('Digite o primeiro número: '))
#input do segundo número do usuário
num2 = int(input('Digite o segundo número: '))

#impressão da divisão normal dos números do usuário
print(num1 / num2)
#impressão da divisão truncada, ou seja, que arredonda o número quebrado
print(num1 // num2)
```

# Exercício 2.

Escreva um programa que lê dois valores em **ponto flutuante** e exiba os seguintes resultados:

* divisão do primeiro valor pelo segundo valor, sem formatação;
* divisão do primeiro valor pelo segundo valor, com exatamente seis dígitos depois do ponto

```python
#input do primeiro número do usuário
n1 = float(input('digite o primeiro número '))

#input do segundo número do usuário
n2 = float(input('digite o segundo número '))

#variável do total da divisão do n1 e n2
total = n1 / n2

#imprime ao usuário a variável total
print(f'Divisão: {total}')

#imprime ao usuário o valor da divisão porém com 6 digitos após o ponto,
#ou seja, formatado
print(f'Divisão formatada: {total:,.6f}')
```

# Exercício 3.

Calcule e apresente o volume de uma lata de óleo dados a altura e o diâmetro da lata.

Lembre: $V = \pi  r^2 A $

```python
#import da biblioteca de matemática
import math

#variável do diametro, obtido pelo usuário
diametro = float(input('Digite o diametro: '))
#variável da altura, obtido pelo usuário
altura = float(input('Digite a altura: '))

#variável do raio, valor necessário para a fórmula
r = diametro / 2

#variável com o valor de pi
pi = math.pi

#fórmula do volume do cilindro
V = pi * (r**2) * altura

#imprime ao usuário o valor do volume do cilindro, formatando-o
print('O volume do cilindro é: {:.2f}'.format(V), 'cm³')
```

# Exercício 4.

Escreva um programa que receba três números inteiros quaisquer e apresente:

* a soma dos quadrados dos três números;
* o quadrado da soma dos três números.


```python
#inputs do primeiro, segundo e terceiro número do usuário
n1 = int(input('Digite o primeiro número: '))
n2 = int(input('Digite o segundo número: '))
n3 = int(input('Digite o terceiro número: '))

#variável que soma os quadrados dos respecivos números
soma = (n1 ** 2) + (n2 ** 2) + (n3 ** 2)

#variável que eleva ao quadrado a soma
somaQ = soma ** 2

#imprime ao usuário a variável soma e somaQ
print(f'A soma dos quadrados dos seus números é: {soma}')
print(f'O quadrado da soma é: {somaQ}')
```
