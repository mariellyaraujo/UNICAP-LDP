# Exercício 1

Faça um programa, utilizando funções, que receba três números inteiros e positivos, e que forneça a soma desses três números.

Para este exercício crie três funções:

> `entrada()` retorna um número digitado (fazer a verificação se é positivo);

> `calculaSoma(a, b, c)` recebe 3 números inteiros e positivos e retorna a soma deles;

> `main()` chamada das funções criadas (chama 3 vezes a entrada, sendo uma para cada número e a função para somar) e depois mostre o resultado.






```python
#funçao da entrada
def entrada():

  #estrutura de repetição que vai repetir até o usuário digitar um valor válido
  while True:

   #input do usuário tipo int
   num = int(input('Digite um número inteiro e positivo: '))

   #caso o número for maior que zero retornar num
   if num > 0:
    return num

   #caso contrário, imprimir o print
   else:
    print('Insira um número positivo')

#função que vai calcular a soma dos 3 números do usuário
def calculaSoma(a, b, c):

  #variável que soma e armazena os números
  total = a + b + c

  #imprime a variável total
  print(total)

#função main
def main():
  #variável que chama a função entrada
  n1 = entrada()
  #variável que chama a função entrada
  n2 = entrada()
  #variável que chama a função entrada
  n3 = entrada()

  #chamada da função calculaSoma com os valores n1, n2 e n3
  calculaSoma(n1, n2, n3)

#chamada da função main
main()
```

# Exercício 2

Faça um programa, utilizando funções, que terá um número inteiro como entrada  e a função retornará o caractere "P", se o número for positivo (> 0),  "N", se o número for negativo (< 0) e "Zero" se o número for igual a zero.

Para este exercício crie duas funções:

> `verifica(a)` recebe o número e retorna "P" (> 0), "N" (< 0) ou "Zero" (igual a 0).

> `main()` digita um número, faz a chamada à função verifica e depois mostra o resultado.






```python
# Função que verifica se um número é zero, positivo ou negativo
def verifica():
    # Solicita ao usuário que digite um número inteiro
    a = int(input('Digite um número: '))

    # Verifica se o número é igual a zero
    if a == 0:
        # Imprime 'Zero' se a condição for verdadeira
        print('Zero')

    # Verifica se o número é maior que zero
    elif a > 0:
        # Imprime 'P' se a condição for verdadeira (número positivo)
        print('P')

    # Se nenhuma das condições anteriores for verdadeira, então o número é negativo
    else:
        # Imprime 'N' se a condição for verdadeira (número negativo)
        print('N')

# Função principal que controla o fluxo do programa
def main():
    # Chama a função verifica para executar a lógica
    verifica()

# Chamada da função principal para iniciar o programa
main()
```

# Exercício 3

Faça um programa, utilizando funções, que retorne o valor de um produto já com o imposto. Serão utilizados dois parâmetros formais: taxaImposto, que é a quantia de imposto sobre vendas expressa em porcentagem e custo, que é o custo de um item antes do imposto.

Para este exercício crie três funções:

> `entrada()` serve para retornar tanto  o custo do produto quanto a porcentagem do imposto;

> `somaImposto(porc, custo)` recebe o valor da porcentagem do imposto e o custo do produto. Retorna o novo custo do produto já com o imposto.

> `main()` chamada das funções criadas (chama 2 vezes a entrada e 1 vez a função somaImposto) e depois mostre o custo com o imposto.






```python
# Função que solicita a entrada de um número inteiro
def entrada():
    # Solicita ao usuário um valor inteiro
    a = int(input())
    # Retorna o valor inserido pelo usuário
    return a

# Função que calcula e exibe o valor do imposto sobre um produto
def somaImposto(a, b):
    # Converte a porcentagem para um valor decimal
    p = a / 100
    # Calcula o total do imposto sobre o valor do produto
    total = b * p
    # Exibe o valor do imposto calculado
    print(total)

# Função principal que controla o fluxo do programa
def main():
    # Solicita ao usuário o valor da porcentagem do imposto
    print('digite o valor da porcentagem do produto:')
    porcen = entrada()  # Chama a função entrada para obter a porcentagem

    # Solicita ao usuário o valor do produto
    print('digite o valor do produto:')
    custo = entrada()  # Chama a função entrada para obter o custo do produto

    # Chama a função somaImposto com a porcentagem e o custo do produto
    somaImposto(porcen, custo)

# Chamada da função principal para iniciar o programa
main()
```

# Exercício 4
Considere o problema de conversão de temperatura:

$$ C = (F - 32) / 1.8 $$

$$ F = 1.8 C + 32 $$



Escreva um programa modularizado que permite ao usuário converter uma faixa de temperatura de Fahrenheit para Celsius (para isso o usuário deve digitar F) e de Celsius para Fahrenheit (neste caso o usuário deve digitar C).

Para a construção do programa você deve escrever as seguintes funções:

> `exibeMsg()` apenas exibe uma mensagem para ao usuário dizendo o que o programa faz e informando como deve ser a entrada de dados. Não tem parâmetro de entrada e não tem retorno;

> `verificaOpcao()` a função não tem parâmetro de entrada e retorna “F” ou “C” (fazer a validação para que o usuário só digite F ou C);

> `verificaIntervalo()` a função não tem parâmetro de entrada e retorna os valores inicial e final do intervalo (fazer a validação para que o valor inicial seja menor que o valor final);

> `exibeFahrenheitToCelsius(inicio, fim)` essa função recebe como entrada o intervalo de temperatura a ser exibido, faz a conversão de temperatura e mostra a temperatura convertida para Celsius;

> `exibeCelsiusToFahrenheit(inicio, fim)` essa função recebe como entrada o intervalo de temperatura a ser exibido, faz a conversão de temperatura e mostra a temperatura convertida para Fahrenheit

```python
# Função que exibe uma mensagem inicial ao usuário
def exibeMsg():
    # Exibe uma mensagem explicativa sobre o que o programa faz
    print("Este programa converte temperaturas entre Celsius e Fahrenheit.")
    print("Digite 'F' para converter de Fahrenheit para Celsius.")
    print("Digite 'C' para converter de Celsius para Fahrenheit.")

# Função que verifica a opção escolhida pelo usuário
def verificaOpcao():
    # Estrutura de repetição que continua até que o usuário digite uma opção válida
    while True:
        # Solicita a entrada do usuário e converte para maiúsculas
        opcao = input("Escolha a opção (F/C): ").strip().upper()
        # Se a opção for 'F' ou 'C', retorna a opção
        if opcao in ('F', 'C'):
            return opcao
        else:
            # Caso contrário, informa que a opção é inválida
            print("Opção inválida! Por favor, digite 'F' ou 'C'.")

# Função que verifica o intervalo de temperatura
def verificaIntervalo():
    # Estrutura de repetição que continua até que o intervalo seja válido
    while True:
        try:
            # Solicita o valor inicial do intervalo
            inicio = float(input("Digite o valor inicial do intervalo: "))
            # Solicita o valor final do intervalo
            fim = float(input("Digite o valor final do intervalo: "))
            # Verifica se o valor inicial é menor que o valor final
            if inicio < fim:
                return inicio, fim
            else:
                # Caso contrário, informa que o intervalo é inválido
                print("O valor inicial deve ser menor que o valor final.")
        except ValueError:
            # Caso ocorra um erro de conversão, informa que a entrada é inválida
            print("Entrada inválida! Por favor, digite números válidos.")

# Função que converte e exibe temperaturas de Fahrenheit para Celsius
def exibeFahrenheitToCelsius(inicio, fim):
    print("\nConvertendo de Fahrenheit para Celsius:")
    # Loop que percorre o intervalo especificado
    for f in range(int(inicio), int(fim) + 1):
        # Calcula a temperatura em Celsius
        c = (f - 32) / 1.8
        # Exibe a conversão
        print(f"{f} °F = {c:.2f} °C")

# Função que converte e exibe temperaturas de Celsius para Fahrenheit
def exibeCelsiusToFahrenheit(inicio, fim):
    print("\nConvertendo de Celsius para Fahrenheit:")
    # Loop que percorre o intervalo especificado
    for c in range(int(inicio), int(fim) + 1):
        # Calcula a temperatura em Fahrenheit
        f = 1.8 * c + 32
        # Exibe a conversão
        print(f"{c} °C = {f:.2f} °F")

# Função principal que controla o fluxo do programa
def main():
    # Chama a função que exibe a mensagem inicial
    exibeMsg()

    # Verifica a opção escolhida pelo usuário
    opcao = verificaOpcao()

    # Verifica e obtém o intervalo de temperatura
    inicio, fim = verificaIntervalo()

    # Dependendo da opção, chama a função apropriada para conversão
    if opcao == 'F':
        exibeFahrenheitToCelsius(inicio, fim)  # Converte de Fahrenheit para Celsius
    else:
        exibeCelsiusToFahrenheit(inicio, fim)  # Converte de Celsius para Fahrenheit

# Chamada da função principal para iniciar o programa
if __name__ == "__main__":
    main()
```
