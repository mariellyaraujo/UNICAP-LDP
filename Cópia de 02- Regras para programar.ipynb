### Questionário: Regras e Boas Práticas de Programação





1. **Por que é importante escrever comentários no código?**
   - A) Para aumentar o tamanho do arquivo.
   - B) Para que o código seja mais facilmente entendido por outras pessoas ou pelo próprio programador no futuro.
   - C) Para tornar o código mais complexo.


```python
#Resposta--> B
```

2. **Qual é a importância de escolher nomes significativos para variáveis e funções?**
   - A) Facilitar a compreensão do código, tornando-o mais legível e autoexplicativo.
   - B) Fazer o código parecer mais profissional.
   - C) Economizar espaço no código.


```python
#Resposta--> A
```


3. **O que significa seguir a prática de "Don't Repeat Yourself" (DRY) em programação?**
   - A) Evitar repetir o código desnecessariamente, encapsulando lógica repetida em funções ou métodos.
   - B) Repetir o código em diferentes partes do programa para garantir que ele funcione.
   - C) Escrever o código uma vez e nunca mais revisá-lo.



```python
#Resposta--> A
```

4. **Por que devemos evitar "hard coding" (inserir valores fixos no código)?**
   - A) Para tornar o código mais flexível e fácil de manter, permitindo a modificação de valores sem alterar o código-fonte.
   - B) Para economizar tempo na escrita do código.
   - C) Para manter o código mais simples.



```python
#Resposta--> A
```

5. **O que é o conceito de modularidade em programação?**
   - A) Dividir o código em partes menores e independentes (módulos), cada uma responsável por uma funcionalidade específica.
   - B) Escrever todo o código em um único arquivo.
   - C) Criar múltiplas cópias do mesmo código.



```python
#Resposta--> A
```

6. **Por que é importante manter um estilo de código consistente, como o PEP 8 em Python?**
   - A) Para garantir que o código seja fácil de ler e manter, promovendo uma uniformidade que facilita a colaboração.
   - B) Para tornar o código mais complexo.
   - C) Para reduzir o tamanho do código.



```python
#Resposta--> A
```

7. **O que é a depuração (debugging) e por que ela é importante?**
   - A) O processo de identificar e corrigir erros no código, fundamental para garantir que o programa funcione corretamente.
   - B) Um método para escrever código mais rápido.
   - C) Uma prática usada para evitar a escrita de testes.



```python
#Resposta--> A
```

8. **Qual é a importância de testar o código regularmente durante o desenvolvimento?**
   - A) Para identificar e corrigir erros cedo, evitando problemas maiores e economizando tempo no longo prazo.
   - B) Para aumentar o tempo de desenvolvimento.
   - C) Para testar apenas no final do projeto.


```python
#Resposta--> A
```


9. **Por que devemos versionar o código, usando ferramentas como Git?**
   - A) Para manter um histórico das mudanças no código, permitindo a colaboração e a reversão para versões anteriores se necessário.
   - B) Para armazenar o código em múltiplos lugares.
   - C) Para aumentar o tamanho do projeto.



```python
#Resposta--> A
```

10. **O que significa "manter o código simples" (KISS - Keep It Simple, Stupid)?**
    - A) Escrever código de forma simples e direta, evitando complexidade desnecessária para facilitar a manutenção e a compreensão.
    - B) Escrever o código de maneira complexa para mostrar suas habilidades.
    - C) Ignorar a simplicidade e focar em otimizações prematuras.

```python
#Resposta--> A
```
