# **Lista**

Uma lista é uma coleção ordenada e mutável. Em Python, as listas são escritas com colchetes.

Nativamente não temos arrays, ou matrizes em Python! Embora possamos empregar, por exemplo, a biblioteca `Numpy` para implementar e operar dados como esses de modo eficiente.

# Descrição dos Métodos


1. **append()**	Adds an element at the end of the list
2. clear()	Removes all the elements from the list
3. copy()	Returns a copy of the list
4. **count()**	Returns the number of elements with the specified value
5. extend()	Add the elements of a list (or any iterable), to the end of the current list
6. **index()**	Returns the index of the first element with the specified value
7. insert()	Adds an element at the specified position
8. pop()	Removes the element at the specified position
9. **remove()**	Removes the item with the specified value
10. reverse()	Reverses the order of the list
11. **sort()**	Sorts the list


# Exercício 1.
Faça um programa que leia uma lista de 5 números inteiros e mostre-os.

> a. Sem função

> b. Com função e variável global

> c. Com função, sem variável global (a função retorna uma lista)


```python
#definição da lista
numeros = []

#for que repete 5 vezes
for i in range(5):
  #pedindo um número ao usuário
  num = int(input("Digite um número: "))
  #adiciona o número do usuário na lista
  numeros.append(num)

#imprime a lista
print(numeros)
```

```python
#definição da lista
numeros = []

#função que lê o número do usuário
def lerNumeros():
  #definição que permite modificar a lista
  global numeros
  #for que repete 5 vezes
  for i in range(5):
    #pedindo um número ao usuário
    num = int(input("Digite um número: "))
    #adiciona o número do usuário na lista
    numeros.append(num)

#função que mostra os componentes da lista
def mostrarLista():
  #for que mostra os componentes da lista
  for numero in numeros:
    print(numero)

#função main
def main():
  #chamada da função lerNumeros
  lerNumeros()
  #chamada da função mostrarLista
  mostrarLista()

#chamada da função main
main()
```

# Exercício 2

Faça um programa que leia uma lista de 10 números inteiros e mostre-os na ordem inversa da digitação.

> a. Sem função

> b. Com função e variável global

> c. Com função, sem variável global (a função retorna uma lista)




```python
# definição da lista
numeros = []

# for que repete 10 vezes
for i in range(10):
    # pedindo um número ao usuário
    num = int(input("Digite um número: "))
    # adiciona o número do usuário na lista
    numeros.append(num)

# reverte a lista de números
numeros.reverse()

# imprime a lista revertida
print(numeros)
```

```python
#definição da lista
numeros = []

#função que lê o número do usuário
def lerNumeros():
  #definição que permite modificar a lista
  global numeros
  #for que repete 10 vezes
  for i in range(10):
    #pedindo um número ao usuário
    num = int(input("Digite um número: "))
    #adiciona o número do usuário na lista
    numeros.append(num)

  # reverte a lista de números
  numeros.reverse()

#função que mostra os componentes da lista
def mostrarLista():
  #for que mostra os componentes da lista
  for numero in numeros:
    print(numero)

#função main
def main():
  #chamada da função lerNumeros
  lerNumeros()
  #chamada da função mostrarLista
  mostrarLista()

#chamada da função main
main()
```

```python
# função que lê o número do usuário
def lerNumeros():
    # definição da lista
    numeros2 = []
    # for que repete 10 vezes
    for i in range(10):
        # pedindo um número ao usuário
        num = int(input("Digite um número: "))
        # adiciona o número do usuário na lista
        numeros2.append(num)
    # reverte a lista de números
    numeros2.reverse()

    # retorna a lista numeros
    return numeros2

# função que mostra os componentes da lista
def mostrarLista(numeros):
    # for que mostra os componentes da lista
    for numero in numeros:
        print(numero)

# função main
def main():
    # chamada da função lerNumeros e armazena o retorno
    numeros = lerNumeros()
    # chamada da função mostrarLista
    mostrarLista(numeros)

# chamada da função main
main()
```

```python
# função que lê quatro notas do usuário
def lerNumeros():
    # definição da lista para armazenar as notas
    notas = []
    # loop que se repete 4 vezes para coletar as notas
    for i in range(4):
        while True:  # loop contínuo até que uma nota válida seja fornecida
            # pedindo uma nota ao usuário
            num = int(input("Digite uma nota de 0 a 10: "))
            # verifica se a nota está fora do intervalo permitido
            if num < 0 or num > 10:
                print("Tente novamente")  # mensagem de erro se a nota for inválida
            else:
                # adiciona a nota válida à lista
                notas.append(num)
                break  # sai do loop se a nota for válida
    # retorna a lista de notas
    return notas

# função que calcula e mostra a média das notas
def mediaNotas(notas):
    # soma das quatro notas
    soma = notas[0] + notas[1] + notas[2] + notas[3]
    # cálculo da média
    media = soma / 4
    # imprime a média das notas
    print("A média das notas:", media)

# função principal
def main():
    # chamada da função lerNumeros e armazena o retorno na variável notas
    notas = lerNumeros()
    # imprime as notas digitadas
    print("Notas digitadas:", notas)
    # chamada da função mediaNotas para calcular e mostrar a média
    mediaNotas(notas)

# chamada da função main para iniciar o programa
main()
```
